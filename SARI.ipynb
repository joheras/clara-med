```python
from datasets import load_dataset
```

```python
dataset = load_dataset("lcampillos/CLARA-MeD")
```

```python
test_set = dataset["train"].shuffle(seed=5)[int(3800*0.8):]
```

mt5-small-clara-med

```python
from transformers import pipeline
```

```python

summarizer = pipeline("text2text-generation", model='joheras/mt5-small-clara-med')
```

```python
from evaluate import load
sari = load("sari")
```

```python
sources = test_set['SOURCE']
predictions = summarizer(test_set['SOURCE'])
predictions = [x['generated_text'] for x in predictions]
references = [[x] for x in test_set['TARGET']]
sari_score = sari.compute(sources=sources, predictions=predictions, references=references)
print(sari_score)
```
